```python
# Ensure plots appear inline in Jupyter
%matplotlib inline

# Core Qiskit imports
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
# Create a 2-qubit quantum circuit
qc = QuantumCircuit(2)

# Apply gates
qc.h(0)      # Hadamard on qubit 0
qc.x(1)      # Pauli-X (NOT) on qubit 1
qc.cx(0, 1)  # CNOT to entangle

# Draw circuit diagram (suppress duplicate output)
_ = qc.draw('mpl')

```

```python
# Simulate the statevector
state = Statevector.from_instruction(qc)
print("Statevector:\n", state)

```

```python
# Convert amplitudes to approximate counts (1024 shots)
counts = {k: int(abs(v)**2 * 1024) for k, v in state.to_dict().items()}
print("Approximate Counts:", counts)

# Plot histogram
plot_histogram(counts)
plt.show()
```

# Author's Note:
1. **Expectation:** Qubits are now in superposition and entangled.
2. **Observation:** Histogram shows approximate probabilities of measured states.
3. **Learning:** Statevector simulation allows us to see quantum amplitudes without needing Aer or execution backends.
