```python
# Ensure plots appear inside Jupyter
%matplotlib inline

# Core Qiskit imports
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_state_city
import matplotlib.pyplot as plt
```

```python
# Create a 2-qubit quantum circuit
qc = QuantumCircuit(2)

# Apply Hadamard gate on first qubit to create superposition
qc.h(0)

# Apply CNOT gate to entangle qubits (Bell state)
qc.cx(0, 1)

# Draw circuit diagram (suppress duplicate output)
_ = qc.draw('mpl')
```

```python
# Get the statevector
state = Statevector.from_instruction(qc)

# Print the statevector
print("Statevector:\n", state)
```

```python
# Bloch sphere for each qubit individually
# Note: For entangled qubits, individual Bloch spheres may look mixed
plot_bloch_multivector(state)
plt.show()

# Optional: 3D city plot for full 2-qubit state
plot_state_city(state)
plt.show()
```

# Author's Note:
1. **Expectation:** Qubits should now be in the Bell state |00> + |11>.
2. **Observation:** Statevector shows equal amplitudes for |00> and |11>, none for |01> or |10>.
3. **Learning:** Entanglement creates correlations that cannot be explained classically; measuring one qubit immediately affects the other.
