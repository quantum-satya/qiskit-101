```python
# Ensure plots appear inside Jupyter
%matplotlib inline

# Core Qiskit imports
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt
```

```python
# Create 1-qubit circuit
qc = QuantumCircuit(1)

# Apply Hadamard gate to create superposition
qc.h(0)

# Visualize circuit
#qc.draw('mpl')
# Draw circuit diagram (suppress duplicate output)
_ = qc.draw('mpl')
```

```python
# Obtain the statevector of the qubit
state = Statevector.from_instruction(qc)

# Plot Bloch sphere
plot_bloch_multivector(state)
plt.show()
```

# Author's Note:
1. **Expectation:** After H gate, the qubit should be in equal superposition |0> + |1>.
2. **Observation:** Bloch sphere shows vector pointing along X-axis  confirms superposition.
3. **Learning:** Qubits can exist in combinations of 0 and 1; Bloch sphere is a powerful visualization tool.
